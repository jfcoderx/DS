### BETWEEN, IN va LIKE

Between biror oraliqdagi malumotlarni chiqarishni sorasak:

```python
# Uzbekistanning faqat country, year, coal, gas, hydro ustunlari 2010 - 2015 oraliqlarida:
query = "SELECT country, year, coal, gas, hydro FROM elect WHERE country='Uzbekistan' AND year BETWEEN 2010 AND 2015"
pd.read_sql(query, db)
```

IN bilan esa istalgan malumotlarni tanlashimiz mumkin:<br>
Misol, 2002, 2005, 2008 malumotlari

```python
# Uzbekistanning faqat country, year, coal, gas, hydro ustunlari 2002, 2005, 2008 dagi malumotlari:
query = "SELECT country, year, coal, gas, hydro FROM elect WHERE country='Uzbekistan' AND year in (2002, 2005, 2008)"
pd.read_sql(query, db)
```

Agar in ichidagi qiymat bolmasa shunchaki chiqarmaydi:

```python
# Uzbekistanning faqat country, year, coal, gas, hydro ustunlari 2002, 2005, 2008, 2020 dagi malumotlari:
query = "SELECT country, year, coal, gas, hydro FROM elect WHERE country='Uzbekistan' AND year in (2002, 2005, 2008, 2020)"
pd.read_sql(query, db)
```

LIKE - biror jihati oxshasa chiqar degani:<br>

% va _

%ar --> Anvar, Sarvar<br>
_ar --> car, bar, tar<br>
__ar --> star, <br>
%ol% --> cOLa, bOLa, bOLalar <br>

```python
# code A dan boshlanadigan davlatlar
query = "SELECT * FROM elect WHERE code LIKE 'A__'"
pd.read_sql(query, db)
```

```python
# oxiri stan bilan tugaydigan davlatlar
query = "SELECT * FROM elect WHERE country LIKE '%stan'"
pd.read_sql(query, db)
```
