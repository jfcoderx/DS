```python
x = np.arange(0, 4*np.pi, 0.1) # 0dan 4pi gacha 0.1 qadam bilan
y = np.sin(x)
```

```python
plt.plot(x, y)
plt.show()
```

qoshimcha argumentlari bilan chiziqlarin ham ozgartirsak boladi:

```python
plt.plot(x, y, linewidth=6)
plt.show()
```

```python
plt.plot(x, y, linewidth=6, linestyle="--")
plt.show()
```

```python
plt.plot(x, y, linewidth=3, linestyle="-.")
plt.show()
```

```python
plt.plot(x, y, linestyle=":", color='r') # red
plt.show()
```

```python
plt.plot(x, y, linestyle="--", color='#8E44AD') # ranlar kodi
plt.show()
```

```python
# bu nuqtalar, har nuqlar toxtashini korsatish uchun markerdan foydalanish mumkin

plt.plot(x, y, marker='.' ) # nuqtalarni toxtash joyini korsatib ketadi
plt.show()
```

```python
plt.plot(x, y, marker='<' ) # nuqtalarni toxtash joyini korsatib ketadi
plt.show()
```

```python
plt.figure(figsize=(16,6)) # grafigimizni hajmini ozgartirish
plt.plot(x, y, marker='o' )
plt.title("sin(x) funksiyasi") # grafiga nom berish
plt.xlabel("PI") # x oqiga nom berish
plt.ylabel("sin(x)") # y oqiga nom berish
plt.grid(color='r', linestyle='--') # grafikka setka urish
plt.show()
```

2 ta grafigni ustma ust chizaylik:

```python
x = np.arange(0, 4*np.pi, 0.1)

sinus = np.sin(x) # sin(x)
cosinus = np.cos(x) # cos(x)

plt.figure(figsize=(12, 6))

# sin(x) grafigi:
plt.plot(x, sinus,
         linewidth=1.5,
         color='r',
         alpha=0.7,   # shaffofligi
         marker='h')

# cos(x) grafigi:
plt.plot(x, cosinus,
         linewidth=1.5,
         color='y',
         alpha=0.7,   # shaffofligi
         marker='h')

# Grafik nomi
plt.title('sin(x) va cos(x) funksiya grafigi')

# x o'q nomi:
plt.xlabel('pi')

# Har biri chiziq haqida malumot, birinchi qaysi chizilsa royxatga shuni yozish kk
# loc - location yani shu malumot qayerda turishi kerakligi
plt.legend(['sin(x)', 'cos(x)'], loc=0, fontsize='x-large')

# setkani ozgartirish
plt.grid(color='g', alpha=0.5, linestyle='--') 

# grafiga ismi va formatini yozib saqlab olish uchun
plt.savefig('grafika.png') 
plt.show()
```

**Grafiklarning turli parametrlarini<br> 
o'zgartirish orqali tashqi ko'rinishini o'zgartirish:**

- `linewidth` - chiziq qalinligi (odatta 0.5 dan 3 gacha berishadi)
- `linestyle` - chiziq stili
    - `-` oddiy chiziq
    - `--` yuliq chiziq
    - `-.` chiziq va nuqta
    - `:` nuqtali chiziq
    - `None` yoki `''` chiziq yo'q
- `color` - chiziq rangi
    - `'b'` - ko'k (blue)
    - `'c'` - havorang (cyan)
    - `'g'` - yashil (green)
    - `'k'` - qora (black)
    - `'m'` - pushti (magenta)
    - `'r'` - qizil (red)
    - `'y'` - sariq (yellow)
    - `HEX ('#FF5733')`
    - `RGBA (255,182,193,0.5)`
    -  https://htmlcolorcodes.com/ turli ranglar mavjud
- `alpha` - chiziq shaffofligi (0 shaffof, 0.5 yarim-shaffof, 1 shaffof emas)
- `marker` - marker
    - `'.'` - nuqta
    - `'o'` - aylana
    - `'v'` - pastga qaragan uchburchak
    - `'^'` - tepaga qaragan uchburchak
    - `'*'` - yulduz
    - To'liq ro'yxat: https://matplotlib.org/3.1.0/gallery/lines_bars_and_markers/marker_reference.html
