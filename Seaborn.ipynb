```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns 

# colabda seabornlar ornatilgan,
# agar ornatilmagan bolsa, pip install seaborn
```

```python
# AQSHdagi bazi muzeylar va ularga turli sanalarda tashrif buyurgan odamlar soni korsatuvchi df

df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/museum_visitors.csv", index_col=0)
df.head()
```

Agar parse_dates ni qoshsak:

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/museum_visitors.csv", index_col=0, parse_dates=[0], usecols=[0, 1])
df.head()
```

```python
type(df.index)
```

Ana endi type datetime boldi, bundan biz oylarni, kunlarni ajratib olishimiz mumkin:

Endi df ga oy va kun ustunlarini qoshamiz:

```python
# df['Year'] = [ind.year for ind in df.index]

df['Year'] = df.index.year
df
```

```python
# df['Month'] = [ind.month for ind in df.index]

df['Month'] = df.index.month
df
```

Biz yil va oy boyicha tashriflarni ajratib olmoqchimiz:

```python
df2 = df.groupby(['Month', 'Year']).max()
df2
# ozi bizda bitta qiymat bor xolos
# togridan togri groupby ni chaqira olmaganimiz uchun max dedik
```

Bizda irarxik indek paydo boldi:<br>
Birinchi indeksi oyni korsatyapti, ikkinchi indeksi yilni korsatyapti

groupby ni aksi bolgan funksiya UNSTACK foydalanamiz<br>
bu nima qiladi, birinchi indeksi ustunga, ikkinchi indeksi ustunga yoyadi<br>
qaysini belgilashimiz, 0 va 1 larni orniga bogliq boladi

```python
df2.unstack(level=0)

# bu yerda qator yillar, ustun oylar
```

```python
df3 = df2.unstack(level=1)
df3
# bu yuqoridagini teskarisi
```

```python
# df3 asosida heatmap chizamiz

plt.figure(figsize=(12,6))
sns.heatmap(data=df3)
plt.show()
```

Ranglar qanday tashriflarni ifodalashi ham korsatilgan:<br>
qanchalik quyuq bolsa shuncha tashriflar kopligini korish mumkin:<br>

Ranglarni ozgartirishimiz ham mumkin:

```python
plt.figure(figsize=(12,6))
sns.heatmap(data=df3, cmap='YlGnBu', linewidths=1)
plt.show()
```

yillar va oylar orni almashtirib kvadrat shaklida chizaylik:

```python
plt.figure(figsize=(12,6))
sns.heatmap(data=df2.unstack(0), cmap='YlGnBu', linewidths=1, square=True)
plt.show()
```

bu yerda oylar orni xunukroq chiqdi:

```python
plt.figure(figsize=(12,6))
sns.heatmap(data=df2.unstack(0), cmap='YlGnBu', linewidths=1, square=True)
xticks_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(np.arange(12) + .5, labels=xticks_labels)
plt.show()
```

### Guruhlangan ustunli grafiklar

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/flight_delays.csv")
df.head()
```

Bu turli oylarda turli aviakompaniyalarga tegishli samolyotlarning
o'rtacha necha minutga<br> kechikkani ko'rsatilgan (har bir ustun bu aviakompaniya nomi qisqartmasi).

manfiy qiymat samolyot erta kelganini anglatadi.

alohida aviakompaniya uchun oyma-oy qiymatlarni ustunli grafik ko'rinishida chiqarishimiz mumkin:

```python
plt.figure(figsize=(16,6))
sns.barplot(x=df['Month'], y=df['AA'])
plt.title("American Airlines kompaniyasi samolyotlari o'rtacha kechikish vaqti")
plt.ylabel("Minutes")
plt.show()
```

En bir nechta aviakompaniyalarni yonma-yon solishtiraylik.<br>
Buning uchun jadvalimizni shaklini o'zgartirishimiz kerak.

Pandasda df shaklini o'zgratirish uchun <br>
`reshape` yoki `melt` funksiyasidan foydalanishimiz mumkin.

`melt` funksiyasi jadvalni eniga uzundan bo'yiga uzun holatga olib kelishda qo'llaniladi. Bunda df dagi<br> bir nechta ustunlarni yagona variable ustuniga keltirish mumkin. Qiymatlar esa har bir variable uchun<br> value ustunida ko'rsatiladi.

Misol koraylik. Bu misolda biz AA va MQ aviakompaniyalari uchun ma'lumotlarni ajratib olganmiz.



```python
df
```

```python
df2 = df.melt(id_vars=['Month'], value_vars=['AA', 'MQ'])
df2
```

Guruhlangan ustunli grafik chizish uchun `catplot` metodidan foydalanamiz `catplot` yordamida<br>
turli grafiklar chizish mumkin, shuning `kind='bar'` parametri yordamida aynan `bar` grafik chizmoqchi<br> ekanmizni ko'rsatamiz.

So'ngra, `x` va `y` o'qi qiymatlarini ko'rsatamiz va eng muhimi guruhlash qaysi ustun bo'yicha bo'lishini<br> ko'rsatish uchun `hue='variable'` parametridan foydalanmiz

```python
plt.figure(figsize=(22,12))
sns.catplot(data=df2, kind="bar", x="Month", y="value", hue="variable")
plt.show()
```

.melt() yordamida istalgancha aviakompaniyalarni ajratib olishimiz mumkin:

```python
df3 = pd.melt(df, id_vars=['Month'], value_vars=['AA', 'AS', 'DL', 'WN', 'MQ'])
df3.head(20)
```

```python
sns.catplot(data=df3, kind="bar", x="Month", y="value", hue="variable")
plt.show()
```

Ko'rib turganingizdek, bunday grafikni tushunish o'ziga yarasha qiyinchilik tug'diradi,<br> bu yerda bizga yana heatmap grafigi yordamga keladi:

```python
plt.figure(figsize=(14,7))
sns.heatmap(data=df.set_index('Month'), cmap="Accent", annot=True)
plt.xlabel("Airline")
plt.show()
```
