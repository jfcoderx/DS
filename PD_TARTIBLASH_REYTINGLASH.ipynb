## SORTING and RANKING

```python
import pandas as pd
import numpy as np
```

```python
s = pd.Series(range(4), index=['d', 'a', 'b', 'c'])
s
```

```python
# Seriesni INDEX si bo'yicha tartiblash

s.sort_index()
```

```python
# Seriesni QIYMATLARI / VALUES bo'yicha tartiblash

s.sort_values()
```

```python
# DataFrame qatorlari yoki ustunlari bo'yicha tartiblash mumkin
# Githubdan kattaroq DF yuklab olamiz
# Amerikadagi sotilayotgan mashinalar jadvali

df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/usa_cars.csv', index_col=0)
df.head()
```

```python
df.shape
```

```python
# Qatorlari yani indekslar boyicha tartiblash
# ammo allaqachon ozi indeks boyicha tartiblangan

df.sort_index()
```

```python
df.sort_index(ascending=False)
```

```python
# Ustunlar nomi bo'yicha tartiblash

df.sort_index(axis=1) 
```

```python
# inplace=True qanday tartiblasak shundayligicha saqlab qoladi

df.sort_index(axis=1, inplace=True) 
df
```

```python
# teskari tartiblash ustunlar boyicha

df.sort_index(axis=1, ascending=False) 
```

```python
# DataFrame elementlarini biror ustun yoki ustunlar bo'yicha tartiblash mumkin

# price qatorni ozi boyicha tartiblash

df.sort_values(by='price')
```

```python
# 1-price qatorni va 2- mileage qatorni mutanosib ravishda TESKARI tartiblash

df.sort_values(by=['price','mileage'], ascending=False)
```

```python
# .RANK() metodi
# df qatorlarning reytingini aniqlash mumkin
# bunda aniq qaysi ko'rsatgich (ustun) bo'yicha aniqlash ko'rsatiladi

# .rank() method parametri yordamida reyting o'lchash usulini 
# tanlash mumkin (min, max, average, first, dense)
```

```python
df['r_year'] = df['year'].rank()
df
```

```python
df.sort_values('year')
```

```python
df['r_year_avr'] = df['year'].rank(method='average')
df.sort_values('year')
```

```python
df.sort_values('year')
```

```python
# Moshinalarni mileage (yurgan kilometri) bo'yicha reyitnglaymiz

df['rank_mileage'] = df['mileage'].rank()
df
```

```python
df.sort_values('mileage')
```

```python
df['r_price_min'] = df['price'].rank(method='min')
df.sort_values('price')
```

```python
df['r_color_avr'] = df['color'].rank(method='average')
df.sort_values('color')
```

```python
df.sort_values(by=['rank_mileage','r_year_avr', 'r_color_avr',  'r_price_min',], ascending=False)
```
