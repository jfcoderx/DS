```python
import pandas as pd
import numpy as np
```

```python
# Series va DataFramelar ustida
# ARIFMETIK AMALLAR
```

```python
s1 = pd.Series([7.3, -2.5, 3.4, 1.5], index=['a', 'c', 'd', 'e'])
s2 = pd.Series([-2.1, 3.6, -1.5, 4, 3.1], index=['a', 'c', 'e', 'f', 'g'])
```

```python
s1
```

```python
s2
```

```python
# Serieslarda arifmetik amallar bevosita indexlar bo'yicha bajariladi
# Bor qiymatni NaN qiymatga qo'shsa, javob NaN chiqaveradi

s1 + s2 
```

```python
d1 = pd.DataFrame(np.arange(9.).reshape(3, 3), columns=list('bcd'), index=['Olma', 'Anor', 'Uzum'])
d2 = pd.DataFrame(np.arange(12.).reshape(4, 3), columns=list('abc'), index=['Olma', 'Anor', 'Qovun', 'Anjir'])
```

```python
d1
```

```python
d2
```

```python
# DataFrame da ARIFMETIK AMALLAR ham qator ham ustun bo'yicha bajariladi
# Ham bir xil QATOR, Ham bir xil USTUN bo'lishi kerak
# Demak NaN qiymatga istalgan qiymat qo'shilsa, NaN hosil bo'ladi

d1 + d2
```

```python
# Qo'shish amalin ADD funksiyasi bilan ham bajarsak bo'ladi
# ADD ning afzalligi, bu funksiya ichida qo'shimcha parametr qo'ya olamiz

# fill_value = 0,
# Bir jadvalda qatori ham bor ustuni ham bor,
# ikkinchi jadvalda esa, ustuni bor yoki qatori bor, yoki ikkalasi ham yoq
# ana shu ikkinchi jadval uchun 0 ni qoyib ketadi

# ammo birini ustuni bor qatori yoq
# ikkinchini qatori yoq ustuni bor bolsa
# bularni qoshganda NaN bo'ladi

d1.add(d2, fill_value=0)
```

```python
# ayirish SUB da qaysini qaysidan ayirish muhim

d2.sub(d1)
```

```python
d2.sub(d1, fill_value=0)
```

```python
d1.sub(d2, fill_value=0)
```

```python
d2.rsub(d1, fill_value=0) # == d1.sub(d2, fill_value=0)
```

```python
d1.div(d2) # == d2.rdiv(d1)
```

```python
d1.rdiv(d2) # == d2.div(d1)
```

```python
df = pd.DataFrame(np.random.randn(3,4), index=list('abc'), columns=['Olma','Anor','Qovun','Anjir'])
df
```

```python
# agar biror funksiyani df ga qollasak, bu funksiya uning barcha elementlariga qo'llaniladi
# modulga olish

np.abs(df)
```

```python
# yaxlitlash

np.round(df)
```

```python
# har bir ustun bo'yicha o'rtacha qiymatni topadi

df.mean()
```

```python
df
```

```python
# YUQORIDAGI AMALLARNI APPLY() METODI YORDAMIDA HAM QO'LLASH MUMKIN
# apply yordami alohid ustun, alohida qatorlarga qo'llash mumkin
```

```python
# alohida ustunlar bo'yicha ortacha qiymat

df.apply(np.mean)
```

```python
# alohida qatorlar bo'yicha ortacha qiymat

df.apply(np.mean, axis=1)
```

```python
df.apply(np.sum)
```

```python
df.apply(np.sum, axis='columns')
```

```python
# APPLY() yana bir afzalligi, istalgan funksiya yozib,
# uni apply orqali, df ning har bir elementiga qo'llashimiz mumkin

func = lambda x: x.max() - x.min()
```

```python
# Demak bunda, ustun bo'yicha hamma elementlarning 
# eng max dan eng min ayiryapti, ular orasidagi farqni ko'rsatyapti

df.apply(func)
```

```python
def foo(x):
  return pd.Series([x.max(), x.min()], index=['min', 'max'])
```

```python
df.apply(foo)
```
