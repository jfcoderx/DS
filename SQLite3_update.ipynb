### SQL jadval yaratish, ma'lumotlar qo'shish

Malumotlar omborida bo'sh jadval yaratish:

```python
query = """
CREATE TABLE IF NOT EXISTS worldpop(
    rank INT, 
    country VARCHAR(50) PRIMARY KEY,
    density DOUBLE,
    population NOT NULL,
    area INT NOT NULL
);
"""
curr.execute(query)
db.commit()
```

query uzun bolsa shunday korinishda qilib olgan yaxshi: <br>
IF NOT EXIST degani agar shunday table mavjud bolmasa yarat degani

`curr.execute(query)` bajarildi va natijasi xotirada turibdi<br>
`db.commit()` shu natijani db ga yozib olinsin degani

Jadval kiritildimi yoki yoqmi yana bir bor tekshirib koraylik:

```python
command = "SELECT name FROM sqlite_master WHERE type='table';"
curr.execute(command)
curr.fetchall()
```

```python
command = "PRAGMA table_info(worldpop);"
curr.execute(command)
curr.fetchall()
```

```python
query = "INSERT INTO worldpop VALUES (1, 'Macau', 21946.4667, 658.3940, 30)"
curr.execute(query)
db.commit()
```

### SQL buyruqlarini f-string yordamida yozish

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/world_population.csv", index_col='rank')
df
```

Yuqoridagi TABLE yana yaratib olamiz:

```python
query = """
CREATE TABLE IF NOT EXISTS worldpop(
    rank INT, 
    country VARCHAR(50) PRIMARY KEY,
    density DOUBLE,
    population NOT NULL,
    
    area INT NOT NULL
);
"""
curr.execute(query)
db.commit()
```

SQL table ga UZBEKISTAN qiymatlarini qaytadan qoshaylik:

```python
uz = df[df.country == 'Uzbekistan']
uz
```

Biz bu qiymatni F STRING orqali SQL TABLE ga qoshamiz

```python
rank = uz.index.values[0]
country = uz.at[rank, 'country']
density = uz.at[rank, 'density']
population = uz.at[rank, 'pop2021']
area = uz.at[rank, 'area']
```

```python
query = f"INSERT INTO worldpop VALUES ({rank}, '{country}', {density}, {population}, {area})"
curr.execute(query)
db.commit()
```

`'{country}'` bu qismini bir tirnoqqa olish kerak<br>
chunki sql querysida string malumot query ichida bir tirnoqqa olingan boladi

```python
curr.execute("SELECT * FROM worldpop")
curr.fetchall()
```

```python
pd.read_sql("SELECT * FROM worldpop", db)   
```
