### Multi Indekslash

Indexlari bir emas bir nechta indekslardan tashkil topgan malumotlar haqida gap ketmoqda

```python
import pandas as pd
import numpy as np
```

Multi indekslashning xususiy holati bu<br> 
`HIERARCHICAL INDEXING` yani IRARXIK INDEKSLASH

Irarxiya tushunchasi bizdagi shachara tushunchasiga togri keladi.<br>

Irarxik indekslash biz qachonki katta malumotlar bilan ishlayotganimizda, malumotlarimizni <br>
guruxlab, indekslarimizni ham guruxlashimizda keng qollanadigan usullardan biri hisoblanadi

```python
data = pd.Series(np.random.randn(9), index=['a', 'a', 'a', 'b', 'b', 'c', 'c', 'd', 'd']) # odatiy indekslash Series
data
```

Indeksini tekshirishaylik:

```python
data.index
```

Bu indeksimizni IRARXIK indeks korinishidagi malumotlarga ogirib olamiz

Bunda biz indeksimizni o'lchamini oshiramiz, ikki o'lchamli qilib

```python
data1 = pd.Series(np.random.randn(9), 
                  index=[['a', 'a', 'a', 'b', 'b', 'c', 'c', 'd', 'd'], [1, 2, 3, 1, 3, 1, 2, 2, 3]]) # odatiy indekslash Series
data1
```

bunda birinchi listdagi 3 ta a, ikkinchi listdagi 1, 2, 3 ni oldi<br>
2 ta b, 1, 3 ni oldi, 2 ta c 1, 2 ni oldi, 2ta d 2, 3 ni oldi

bu kabi usul - yuqori mashtabli malumotlar bilan ishlayotganimizda <br>guruxlashda juda ham qo'l keladi, bu datamizni indeksini tekshiradigan bo'lsak:

```python
data1.index
```

Multi Indeks ga kelib qoldi

Irarxik indekslarni bir qismini ajratib olish, <br>
avvalgi usullardan unchalik farq qilmaydi

```python
data1['a']
```

bunday indekslaganda subsetlari bilan ajratib oladi.

oddiy indekslaganda qanday kesishlar<br>
ishlagan bolsa bunda ham kopi ishlaydi:

```python
data1['b': 'c'] 
```

a, c, d ning 2-indeks malumotlarini<br> 
ajratib oladi, 2-indeksi yoqini ajratib olmaydi

```python
data1.loc[:, 2]
```

barcha b va d indeksdagi elementlarni ajratib olish:

```python
data1.loc[['b', 'd']]
```

### Qayta Tartiblash:

Indekslari yoki ustunlarni qanday qilib bir bir bilan ornini<br>
almashtirishni korib chiqamiz, ularni qayta tartiblashlarni ham:

```python
df3
```

#### swaplevel()

darajalarni ozgartirish degan metoddan foydalanamiz<br>
bu bizga irarxik indekslarning ornini almashtirishdan <br>
imkonini beradi, yani countryni cityga

swaplevel ichida argument sifatida columnlar jadvalda qanday<br>
 holatda turgan bolsa shunday ketma ketlikda yozishimiz kerak:

```python
df3.swaplevel('Country', 'City')
```

Korganingizdek orni almashdi va Country da bir davlatlar bir necha bor kelgan

#### sort_index 

uning avzal jihatlaridan biri, swaplevel() ning ichida ham yozib<br>
 ketaversak boladi, unin asosiy vazifasi indekslarni tartiblashdan iborat

indekslarni teskari tartib chiqaraylik:<br>
buni bajarish uchunam df3 ni chaqirib olamiz:
sort_index() argumenti sifatida level kiritami

level bizga indekslarimizni qaysiligini ifoydalaydi
level = 0 va 1 ikta levelimi 

level=0 orniga nolni kiritadigan bolsak, asl malumotimizdagi level yani darajani qaytargam boladi

agar 1 qiladigan bol'lsa, ularni orni almashadi city bilan country orni almashib qoladi

Biz asl holatga qaytarmoqchimiz;

```python
df3.sort_index(level=1, ascending=True)
```
