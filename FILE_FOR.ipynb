## HDF5 FORMATI

```python
# bu formatni avzalligi boshqa dasturlash tillarida ham ochish mumkinligi
# va aynan ilmiy izlanishlar qilinganda katta fayllarni saqlashda ishlatiladigan
# formatlarda biri, va katta fayllarni saqlaganda jadvallarni siqilgan holatda saqlaydi
```

```python
# bu formatni hosil qilish uchun yangi obyekt yaratib olamiz
# objh5 obyektimizdagi ozgarishlar togridan togri
# datah5 faylimizga saqlanadi, uni formati h5

objh5 = pd.HDFStore('datah5.h5')
```

```python
df
```

```python
# biz bir nechta dataframelarni h5 obyektimizga saqlashimiz mumkin boladi
# faqat biror kalit soz bilan

objh5.put('cardata', df)
objh5.put('calidata', df1)
```

```python
# objh5 ni kalitlarini korishimiz mumkin:

objh5.keys()
```

```python
# lugatdagi kalitlardan malumotlarni chaqirganimiz kabi
# bunda ham kalit sozlari bilan df larni chaqirsak boladi

objh5['cardata']
```

```python
objh5['calidata']
```

```python
# hozir bizni datah5 miz ochiq, barcha ozgarishlarni qilib bolgach
# bu faylimizni yopishimiz kerak

objh5.close()
```
